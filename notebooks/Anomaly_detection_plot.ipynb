```python
plt.figure(figsize=(12,6))

normal = df[df["ANOMALY"] == 1]
anomaly = df[df["ANOMALY"] == -1]

plt.scatter(normal["DATE"], normal["PRECTOTCORR"], color="blue", alpha=0.4)
plt.scatter(anomaly["DATE"], anomaly["PRECTOTCORR"], color="red")

plt.title("Climate Anomaly Detection")
plt.xlabel("Date")
plt.ylabel("Rainfall")
plt.legend(["Normal","Anomaly"])

plt.show()
```
